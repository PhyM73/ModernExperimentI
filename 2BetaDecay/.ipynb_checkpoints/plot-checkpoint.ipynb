```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema as extre
from scipy.optimize import curve_fit as fit

def gauss(x, a, mu, sigma):
    return a*np.exp(-(x-mu)**2/(2.*sigma**2))
```

```python
import os

def loaddata(rootpath, isex=True):
    data = {}
    for path in os.listdir(rootpath):
        if path.startswith('.'): 
            continue
        subpath = rootpath+'/'+path+'/'
        file = [i for i in os.listdir(subpath) if i.endswith('txt')]
        
        # load data, and divide by livetime if it is from experiments
        if isex:
            data[path[:3]] = {i[:-8] : np.genfromtxt(subpath+i,delimiter=' ')[:,1]/int(i[-7:-4]) for i in file}
        else:
            data[path[:3]] = {i[:-4] : np.genfromtxt(subpath+i,delimiter=' ')[:,1] for i in file}
    return data

ex = loaddata('./20201021')
mc = loaddata('./20201021MC', isex = False)
```

```python
channel = np.array(range(512))
```

```python
def smooth(series,strength=1):
    return np.convolve(series, np.ones(2*strength+1)/(2*strength+1), mode='same' )
```

```python
def exbkg(data,left=20,right=290):
    win = smooth(data[left:right+1])
    lvc = np.where(win==np.min(win))[0][0] + left #左边界 left v channel
    pkc = np.where(win==np.max(win))[0][0] + left
    rvc = min(2*pkc-lvc,511) #右边界 right v channel
    print(lvc,rvc)

    lvh = np.average(data[lvc-3:lvc+4])
    rvh = np.average(data[rvc-3:rvc+4])
    delta_rl= lvh-rvh
    sum_rl = np.sum(data[lvc:rvc+1])

    bkg = np.array([np.sum(data[lvc+i:rvc+1])/sum_rl*delta_rl + rvh for i in range(rvc+1-lvc)])
    sig = data[lvc:rvc+1] - bkg
    sig = np.concatenate((np.zeros(lvc),sig,np.zeros(511-rvc)))
    
    return sig, lvc, pkc, rvc
```

```python
def findpeak(sig,lvc,pkc,rvc):
#     sig, lvc, pkc, rvc = exbkg(data,left,right)
    p0= np.array([0,0,0])
    p = np.array([sig[pkc], pkc, (rvc-lvc)/2]) # a, mu, sigma
    var = np.array([0,0,0])
    
    win = channel[np.logical_and(p[1]-p[2]-1<channel,channel<p[1]+p[2]+1)] # window
    
    loops = 0
    while np.any(np.absolute(p-p0) > var) and loops < 8:
        p0 = p 
        popt, pcov = fit(gauss,win,sig[win], p0=p0)
        p = np.absolute(popt)
        var = np.array([np.sqrt(pcov[i,i]) for i in range(3)])
        win = channel[np.logical_and(p[1]-p[2]-1<channel,channel<p[1]+p[2]+1)] # window
        loops += 1
#         print(p,var)
        
    return popt, var 
```

```python
# find peak method illustration

(fig,ax_pk) = plt.subplots()
ax_pk.set_xlabel('Channel')
ax_pk.set_ylabel('Rate ($s^{-1}$)')

data = ex['Vac']['233mm']

sig0, lvc0, pkc0, rvc0 = exbkg(data)
(p,v)=findpeak(sig0,lvc0,pkc0,rvc0)

ax_pk.plot(channel,data,'.',markersize=1,label='rates')
ax_pk.plot(channel,sig0,markersize=1,label='peak')
ax_pk.plot(channel,data-sig0,linestyle='--',label='bkg')
ax_pk.axvline(lvc0,ls='-.',linewidth=1,color='C0')
ax_pk.axvline(pkc0,ls='-.',linewidth=1,color='C0')
ax_pk.axvline(rvc0,ls='-.',linewidth=1,color='C0')

fitl = int(p[1]-p[2])
fitr = int(p[1]+p[2])
ax_pk.plot(channel[fitl:fitr],p[0]*np.exp(-(channel[fitl:fitr]-p[1])**2/(2.*p[2]**2)),linestyle='-',label='fit')


ax_pk.legend()
fig.savefig('pic/illu.pdf')
```

```python
# calibrate
(fig,ax_cs) = plt.subplots()
ax_cs.set_xlabel('Channel')
ax_cs.set_ylabel('Rate (s$^{-1}$)')

sigs = []
peaks = []

Csdata = ex['Cal']['Cs137']
sigs.append(exbkg(Csdata,left=20,right=40))
sigs.append(exbkg(Csdata,left=100,right=160))

Codata = ex['Cal']['Co60']*2
sigs.append(exbkg(Codata,left=240,right=270))
sigs.append(exbkg(Codata,left=280,right=310))

for i in sigs:
    peaks.append(findpeak(*i))

ax_cs.plot(channel,Csdata,'.',markersize=1,label='Cs137 rates')
ax_cs.plot(channel,sigs[0][0]+sigs[1][0], markersize=1,label='Cs137 peak')
ax_cs.axvline(peaks[0][0][1],ls='-.',linewidth=1,color='C4',label='chn %.1f, %.1f'%(peaks[0][0][1],peaks[1][0][1]))
ax_cs.axvline(peaks[1][0][1],ls='-.',linewidth=1,color='C4')

ax_cs.plot(channel,Codata,'.',markersize=1,label='Co60 rates * 2')
ax_cs.plot(channel,sigs[2][0]+sigs[3][0], markersize=1,label='Co137 peak')
ax_cs.axvline(peaks[2][0][1],ls='-.',linewidth=1,color='C4',label='chn %.1f, %.1f'%(peaks[2][0][1],peaks[3][0][1]))
ax_cs.axvline(peaks[3][0][1],ls='-.',linewidth=1,color='C4')

ax_cs.legend()
fig.savefig('pic/cal.pdf')
```

```python
# calibrate MC
(fig,ax_mccal) = plt.subplots()
ax_mccal.set_xlabel('Channel')
ax_mccal.set_ylabel('Rate (s$^{-1}$)')

mcsgs = []
mcpks = []

mcCs = mc['Cal']['Cs137']
mcsgs.append(exbkg(mcCs,left=25,right=50))
mcsgs.append(exbkg(mcCs,left=120,right=160))

mcCo = mc['Cal']['Co60']/4
mcsgs.append(exbkg(mcCo,left=250,right=265))
mcsgs.append(exbkg(mcCo,left=290,right=310))

for i in mcsgs:
    mcpks.append(findpeak(*i))

ax_mccal.plot(channel,mcCs,'.',markersize=1,label='Cs137 counts')
ax_mccal.plot(channel,mcsgs[0][0]+mcsgs[1][0], markersize=1,label='Cs137 peak')
ax_mccal.axvline(mcpks[0][0][1],ls='-.',linewidth=1,color='C4',label='chn %.1f, %.1f'%(mcpks[0][0][1],mcpks[1][0][1]))
ax_mccal.axvline(mcpks[1][0][1],ls='-.',linewidth=1,color='C4')

ax_mccal.plot(channel,mcCo,'.',markersize=1,label='Co60 counts')
ax_mccal.plot(channel,mcsgs[2][0]+mcsgs[3][0], markersize=1,label='Co137 peak')
ax_mccal.axvline(mcpks[2][0][1],ls='-.',linewidth=1,color='C4',label='chn %.1f, %.1f'%(mcpks[2][0][1],mcpks[3][0][1]))
ax_mccal.axvline(mcpks[3][0][1],ls='-.',linewidth=1,color='C4')

ax_mccal.legend()
fig.savefig('pic/mccal.pdf')
```

```python
def gen_1Dlabel(para, cov,x='chn', d=[5,5,3,3], n=[1,1]):
    f = []
    for i in d:
        f.append('{:.'+str(i)+'f}')
    return (f[0]+'({})*'+ x + ' + '+f[2]+'({})').format(para[0],f[1].format(np.sqrt(cov[0][0]))[-n[0]:], 
                                                        para[1],f[3].format(np.sqrt(cov[1][1]))[-n[1]:])
```

```python
# calibrate fit
(fig,ax_cal) = plt.subplots()

ax_cal.set_xlabel('Channel')
ax_cal.set_ylabel('Energy (MeV)')

exchn = [peaks[i][0][1] for i in range(len(peaks))]
mcchn = [mcpks[i][0][1] for i in range(len(mcpks))]
eng = [0.184,0.662,1.173,1.333]
ax_cal.plot(exchn,eng,'+',markersize=12)
ax_cal.plot(mcchn,eng,'+',markersize=12)

scale1, cov1 = np.polyfit(exchn, eng, 1,cov=True)
scale2, cov2 = np.polyfit(mcchn, eng, 1,cov=True)
ax_cal.plot(channel,scale1[0]*channel+scale1[1],label='EX $E$ = ' + gen_1Dlabel(para=scale1,cov=cov1))
ax_cal.plot(channel,scale2[0]*channel+scale2[1],label='MC $E$ = ' + gen_1Dlabel(para=scale2,cov=cov2))

ax_cal.legend()
fig.savefig('pic/calfit.pdf')
```

```python
# al absorb
(fig,ax_al) = plt.subplots()

ax_al.set_xlabel('Channel')
ax_al.set_ylabel('Rates (s$^{-1}$)')
ax_eng = ax_al.twiny()
ax_eng.set_xlabel('Energy (MeV)')

ala_peaks=[[],[]]
ala_pkr=[[],[]] # peak rates

for key, value in ex['Ala'].items():
    width= '%.2f $mm$'%(float(key[3])*0.05)
    (pic,)=ax_al.plot(channel,value,'.',markersize=1 ,label=width+' rates')
    (pic,)=ax_eng.plot(channel*scale1[0]+scale1[1],value,'.',markersize=1 ,label=width+' rates')
    sign = exbkg(value,left=70,right=270)
    ax_al.plot(channel,sign[0],linewidth=1,c=pic.get_color(),label=width+' peak')
    
    peak = findpeak(*sign)
    ala_peaks[0].append(peak[0][1]*scale1[0]+scale1[1])
    ala_peaks[1].append(peak[1][1]*(scale1[0]+np.sqrt(cov1[0][0])) + np.sqrt(cov1[1][1]))
    ala_pkr[0].append(value[int(round(peak[0][1]))])
    ala_pkr[1].append(peak[1][1])

ax_al.legend()
fig.savefig('pic/alabsorb.pdf')
```

```python
# mc al absorb
(fig,ax_almc) = plt.subplots()

ax_almc.set_xlabel('Channel')
ax_almc.set_ylabel('Counts ')
ax_engmc = ax_almc.twiny()
ax_engmc.set_xlabel('Energy (MeV)')

mc_ala_peaks=[[],[]]
mc_ala_pkr=[[],[]]
for key, value in mc['Ala'].items():
    width= key[2:6]+' mm'
    value = value/5000
    (pic,)=ax_almc.plot(channel,value,'.',markersize=1 ,label=width+' counts')
    (pic,)=ax_engmc.plot(channel*scale2[0]+scale2[1],value,'.',markersize=1 ,label=width+' counts')
    mcsign = exbkg(value,left=20,right=260)
    ax_almc.plot(channel,mcsign[0],linewidth=1,c=pic.get_color(),label=width+' peak')

    peak = findpeak(*mcsign)
    mc_ala_peaks[0].append(peak[0][1]*scale2[0]+scale2[1])
    mc_ala_peaks[1].append(peak[1][1]*(scale2[0]+np.sqrt(cov2[0][0])) + np.sqrt(cov2[1][1]))
    mc_ala_pkr[0].append(value[int(round(peak[0][1]))])
    mc_ala_pkr[1].append(peak[1][1])
    
ax_almc.legend()
```

```python
# al absorb energy fit
(fig,ax_alfit) = plt.subplots()
ax_alfit.set_xlabel('Thickness (mm)')
ax_alfit.set_ylabel('Energy (MeV)')

thick = np.array([0,0.05,0.1,0.15,0.2])

k_ala, c_ala = np.polyfit(thick,ala_peaks[0], 1,cov=True)
k_mcala, c_mcala = np.polyfit(thick, mc_ala_peaks[0], 1,cov=True)

datap=ax_alfit.errorbar(thick,ala_peaks[0],xerr=thick*0.05,yerr=ala_peaks[1],fmt='o',markersize=3,capsize=3,label='Data',c='C0')
ax_alfit.plot(thick,k_ala[0]*thick+k_ala[1],label='$E$ = '+gen_1Dlabel(para=k_ala,cov=c_ala,x='d',d=[3,3,4,4]),c='C0')

mcp=ax_alfit.errorbar(thick,mc_ala_peaks[0],xerr=thick*0.05,yerr=mc_ala_peaks[1],fmt='o',markersize=3,capsize=3, label='MC',c='C1')
ax_alfit.plot(thick,k_mcala[0]*thick+k_mcala[1],label='$E$ = '+gen_1Dlabel(para=k_mcala,cov=c_mcala,x='d',d=[3,3,4,4]),c='C1')

ax_alfit.legend()
fig.savefig('pic/absorbenfit.pdf')
```

```python
def gen_elabel(para, cov,x='d', d=[5,5,3,3], n=[1,1]):
    f = []
    for i in d:
        f.append('{:.'+str(i)+'f}')
    return (f[0]+'({})*exp( -'+x+'/'+f[2]+'({}) )').format(np.exp(para[1]),f[1].format(np.exp(para[1])*np.sqrt(cov[1][1]))[-n[0]:], 
                                                           abs(1/para[0]),f[3].format(np.sqrt(cov[0][0])/para[0]**2)[-n[1]:])
```

```python
# al absorb rates fit
(fig,ax_alfite) = plt.subplots()
ax_alfite.set_xlabel('Thickness (mm)')
ax_alfite.set_ylabel('Rates (s$^-1$)')

x = np.linspace(-0.005,0.205,52)

thick = np.array([0,0.05,0.1,0.15,0.2])
A_ala, C_ala = np.polyfit(thick,np.log(np.array(ala_pkr[0])), 1, cov=True)
A_mcala, C_mcala = np.polyfit(thick,np.log(np.array(mc_ala_pkr[0])), 1, cov=True)

datap=ax_alfite.errorbar(thick,ala_pkr[0],xerr=thick*0.05,yerr=ala_pkr[1],fmt='o',markersize=3,capsize=3,label='Data',c='C0')
ax_alfite.plot(x,np.exp(A_ala[0]*x+A_ala[1]),label='$R$ = '+gen_elabel(para=A_ala,cov=C_ala,d=[2,2,2,2],n=[2,2]),c='C0')

mcp=ax_alfite.errorbar(thick,mc_ala_pkr[0],xerr=thick*0.05,yerr=mc_ala_pkr[1],fmt='o',markersize=3,capsize=3, label='MC',c='C1')
ax_alfite.plot(x,np.exp(A_mcala[0]*x+A_mcala[1]),label='$R$ = '+gen_elabel(para=A_mcala,cov=C_mcala,d=[2,2,3,3]),c='C1')

ax_alfite.legend()
fig.savefig('pic/absorbratefit.pdf')
```
